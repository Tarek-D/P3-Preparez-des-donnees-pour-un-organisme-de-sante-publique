```python
# Compter le nombre de valeurs manquantes par colonne
missing_values = data.isnull().sum()

# Calculer le pourcentage de valeurs manquantes par colonne
missing_percentage = (missing_values / len(data)) * 100

# Cr√©er un dataframe pour afficher les statistiques des valeurs manquantes
missing_data = pd.DataFrame({'Missing Values': missing_values, 'Missing Percentage': missing_percentage})
missing_data.sort_values(by='Missing Percentage', ascending=False, inplace=True)

# Afficher les statistiques des valeurs manquantes
display(missing_data)
```

## Suppression des colonnes trop vides, seuil fix√© √† 80%

```python
# Seuil de pourcentage de valeurs manquantes
threshold = 80

# Liste des colonnes √† supprimer
columns_to_drop = missing_data[missing_data['Missing Percentage'] > threshold].index

# Supprimer les colonnes
data.drop(columns=columns_to_drop, inplace=True)
```

## Supprimer les lignes dupliqu√©es

```python
# Supprimer les lignes dupliqu√©es
data.drop_duplicates(inplace=True)
```

```python
# Rep√©rer les variables pertinentes pour les traitements √† venir
variables_pertinentes = ['code', 'product_name', 'quantity', 'packaging', 'categories',
                         'ingredients_text', 'serving_size', 'additives', 'nutrition_grade_fr',
                         'main_category']

# Calculer le nombre de valeurs manquantes pour chaque variable pertinente
valeurs_manquantes = data[variables_pertinentes].isnull().sum()

# Afficher le nombre de valeurs manquantes
print(valeurs_manquantes)
```

## Colonnes avec plus de 50 % de valeurs manquantes

```python
selected_variables = ['code', 'product_name', 'quantity', 'packaging', 'brands', 'main_category', 'ingredients_text', 'serving_size', 'additives_n', 'nutrition_grade_fr']
missing_values = data[selected_variables].isnull().mean()
missing_variables = missing_values[missing_values > 0.5].index
display(missing_variables)
```

### AJOUTER ETAPE ENREGISTREMENT NOUVEAU DATA CLEAN

```python
# Convertir la colonne 'quantity' en type num√©rique
data['quantity'] = pd.to_numeric(data['quantity'], errors='coerce')

# D√©finir les seuils
lower_threshold = data['quantity'].quantile(0.25) - 1.5 * data['quantity'].std()
upper_threshold = data['quantity'].quantile(0.75) + 1.5 * data['quantity'].std()

# Identifier les valeurs aberrantes
outliers = data[(data['quantity'] < lower_threshold) | (data['quantity'] > upper_threshold)]

# Supprimer les valeurs aberrantes
data = data[(data['quantity'] >= lower_threshold) & (data['quantity'] <= upper_threshold)]

# Tracer le diagramme de bo√Æte
plt.figure(figsize=(8, 6))
plt.boxplot(data['quantity'], vert=False)
plt.title('Diagramme de bo√Æte pour quantity')
plt.xlabel('Valeur')
plt.ylabel('quantity')
plt.show()
```

```python
# Afficher les diff√©rentes valeurs uniques et leur fr√©quence
packaging_counts = data['packaging'].value_counts()

# S√©lectionner les 10 premi√®res valeurs d'emballage les plus fr√©quentes
top_10_packaging = packaging_counts.head(20)

# Cr√©er le graphique √† barres
plt.figure(figsize=(10, 6))
top_10_packaging.plot(kind='bar')
plt.title('Top 20 des valeurs d\'emballage les plus fr√©quentes')
plt.xlabel('Emballage')
plt.ylabel('Fr√©quence')
plt.show()
```

```python
# Nettoyer les valeurs (y compris les valeurs manquantes)
data['main_category'] = data['main_category'].str.lower().str.strip()

# Analyse des fr√©quences des cat√©gories
category_counts = data['main_category'].value_counts()
display(category_counts)
```
